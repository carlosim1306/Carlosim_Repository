```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('img/fotito.jpg')
print(img)
```

```python
plt.imshow(img)
plt.show()
```

```python
print('La imgen es una matriz de tamao',img.shape, '(px, px, canales))')
```

```python
print('Esto es el pixel [100, 100]: ',img[100,100])
```

```python
print('Esto es la fila 100: ',img[100, :])
```

```python
print('Esto es una submatriz: ',img[100:200, 100:200])
```

```python
b,g,r = cv2.split(img)
img_rgb = cv2.merge ((r,g,b))
plt.imshow(img_rgb)
plt.show()
```

```python
cola = img[0:100, 300:400]
lenx, leny, ch = cola.shape
print(cola.shape)
img[200:200+lenx, 50:50+leny] = cola
img[200:200+lenx, 250:250+leny] = cola
img[200:200+lenx, 450:450+leny] = cola

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show
```

```python
cv2.imwrite('ej3_fotito_modificada.jpg', img)
```
