```python
import numpy
```

```python
x = numpy.arange(0, numpy.pi, 0.01)
y = numpy.sin(x)
z = numpy.cos(x)
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt

plt.figure(1)
plt.plot(x)
plt.ylabel('Lista de valores x')
plt.show()
```

```python
plt.figure(2)
plt.plot(y)
plt.ylabel('Lista de valores y')
```

```python
plt.figure(3)
plt.plot(x,y)
plt.ylabel('Sin(x)')
plt.ylabel('x')
```

```python
plt.figure(4)
plt.plot(x,y,'r')
plt.plot(x,z,'g')
plt.legend(['Sin(x)', 'Cos(x)'])
plt.xlabel('x')
plt.ylabel('Values')
```
